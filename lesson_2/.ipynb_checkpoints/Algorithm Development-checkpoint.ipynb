```python
# Run once

searcher = bingImagesRetriever(apiKeyName='AZURE_SEARCH_KEY')
searcher.retrieveApiKey()

generalName = 'plant'
specificNamesArray = ['thyme', 'rosemary']
pathName = 'plants'

searcher.downloadImagesToPath(
    generalName = generalName,
    specificNamesArray = specificNamesArray,
    pathName = pathName)
searcher.unlinkFailed(pathName)
```

```python
from fastai import *
from fastai.vision.all import *
```

```python
plants = DataBlock(
    blocks=(ImageBlock, CategoryBlock),
    get_items=get_image_files,
    splitter=RandomSplitter(valid_pct=0.2, seed=42),
    get_y=parent_label,
    item_tfms=Resize(128)
)
```

```python
path = Path(pathName)
dls = plants.dataloaders(path)
```

```python
dls.valid.show_batch(max_n=4, nrows=1)
```

```python
plants = plants.new(
    batch_tfms=aug_transforms(),
    item_tfms=RandomResizedCrop(224, min_scale=0.5)
)

dls = plants.dataloaders(path)
```

```python
# has to be run in a GPU
learn = cnn_learner(dls, resnet18, metrics=error_rate)
learn.fine_tune(4)
```
